```python
list(split['train']['Drug'])
```

```python
smi = list(split['train']['Drug'])[0]
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem
```

```python
smi
```

```python
mol = Chem.MolFromSmiles(smi)
```

```python
atom_list = mol.GetAtoms()
len(atom_list)
```

```python
print(atom_list[0].GetSymbol())
print(atom_list[0].GetDegree())
print(atom_list[0].GetImplicitValence())
print(atom_list[0].GetIsAromatic())
print(atom_list[0].GetHybridization())
print(atom_list[0].GetTotalNumHs())
```

```python
bond_lists = mol.GetBonds()
len(bond_lists)
```

```python
print(bond_lists[0].GetBeginAtom().GetSymbol())
print(bond_lists[0].GetBeginAtom().GetIdx())
print(bond_lists[0].GetEndAtom().GetSymbol())
print(bond_lists[0].GetEndAtom().GetIdx())
print(bond_lists[0].GetBondType())
print(bond_lists[0].IsInRing())
print(bond_lists[0].GetIsAromatic())
print(bond_lists[0].GetIsConjugated())
```
